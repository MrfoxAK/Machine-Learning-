```python
null_var = df.isnull().sum()/df.shape[0] *100
null_var
```

```python
drop_col = null_var[null_var>17].keys()
drop_col
```

```python
df3 = df.select_dtypes(include=['int64','float64'])
df3.head(5)
```

```python
sns.heatmap(df3.isnull())
```

```python
missing_num_var = [var for var in df3.columns if df3[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3[var], bins=20, kde_kws={'linewidth':5, 'color':'#DC143C'})
```

```python
df4 = df3.fillna(df3.mean())
df4.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    plt.legend()
```

```python
df5 = df3.fillna(df3.median())
df5.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df5[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    plt.legend()
```
