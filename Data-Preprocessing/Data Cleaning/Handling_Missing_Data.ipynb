```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isna())
```

```python
null_var = df.isnull().sum()/df.shape[0]*100
```

```python
drop_col = null_var[null_var>10].keys()
drop_col
```

```python
df2_drop_clm = df.drop(columns=drop_col)
```

```python
df2_drop_clm.shape
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df2_drop_clm.isna())
```

```python
df3_drop_rows = df2_drop_clm.dropna()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df3_drop_rows.isna())
```
